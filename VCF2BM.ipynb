## Day 1: Identification of genomic Biomarker for TNBC-01

```python
# Import necessary libraries
import pandas as pd
```

```python
# Load the IntOGen-DriverGenes_BRCA.tsv file into a pandas DataFrame
drivers = pd.read_csv('Day2/TNBC01/Inputs/IntOGen-DriverGenes_BRCA.tsv', sep='\t')
```

```python
# Load the TNBC01.VEP.output.txt file into a pandas DataFrame
df = pd.read_csv('Day2/TNBC01/Inputs/TNBC01.VEP.output.txt', sep='\t')
```

```python
# Replace any '-' values in the 'gnomADe_AF' column with 0 and convert the column to float type
df['gnomADe_AF'] = df['gnomADe_AF'].replace({'-': 0}).astype(float)
df = df.loc[df['gnomADe_AF'] < 0.01]
df = df.loc[df['IMPACT'].isin(['HIGH', 'MODERATE'])]
df = df.loc[df['CANONICAL'] == 'YES']
df = df.loc[df['SYMBOL'].isin(drivers['Symbol'])]
```

```python
# Display the 'SYMBOL' and 'HGVSp' columns of the filtered DataFrame
df[['SYMBOL', 'HGVSp']]
```
